```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("Virat Kohli DataSet Clean.csv")
```

```python
dataset.head(1)
```

```python
dataset.shape[0]
```

```python
dataset[(dataset['Opposition']=="Sri Lanka") & (dataset['Ground']=="Dambulla")]['Pos'].unique()
```

# Q9

```python
dataset['Ground'].value_counts()[:5]
```

```python
plt.figure(figsize=(7,3))
sns.countplot(x="Dismissal",data=dataset[dataset['Ground']=="Dhaka"],color = "red")
plt.ylabel("Count of Match")
plt.title("No of Dismissal in ODI Match Ground Dhaka")
plt.yticks([i for i in range(0,6,1)])
sns.set_style(style="dark")
plt.savefig("No of Dismissal in ODI Match Ground Dhaka.jpg")
plt.show()
```

# Q8

```python
sns.lineplot(x="BF",y="4s",data=dataset,errorbar=('ci', 0),label="4s",color = "blue")
sns.lineplot(x="BF",y="6s",data=dataset,errorbar=('ci', 0),label="6s",color="red")
plt.ylabel("Counts of 4s,6s")
plt.xlabel("Ball Face")
plt.title("Ball Face v/s 4s,6s Correlation")
sns.set_style(style="dark")
plt.savefig("Ball Face vs 4s,6s Correlation.jpg")
plt.show()
```

# Q7

```python
sns.scatterplot(x="BF",y="Runs",data=dataset,color="red")
plt.xlabel("Ball Face")
plt.title("Ball Face v/s Runs Correlation")
sns.set_style(style="dark")
plt.savefig("Ball Face vs Runs Corr.jpg")
plt.show()
```

# Q6

```python
dataset['SR'].mean()
```

```python
dataset['SR'].min() , dataset['SR'].max()
```

```python
sns.histplot(x="SR",data=dataset,bins=[i for i in range(0,220,20)],color="red")
plt.xticks([i for i in range(0,220,20)])
plt.title("Strike Rate in ODI Match")
plt.xlabel("Strike Rate")
plt.ylabel("Count of Match")
sns.set_style(style="dark")
plt.savefig("Strike Rate in ODI Match.jpg")
plt.show()
```

# Q5

```python
sns.barplot(x="Pos",y="Runs",data=dataset,estimator="sum",color="red",errorbar=('ci', 0))
plt.title("Best Position in ODI Match")
plt.xlabel("Position")
sns.set_style(style="dark")
plt.savefig("Best Position in ODI Match.jpg")
plt.show()
```

# Q4

```python
x = dataset['Inns'].value_counts().index
y = dataset['Inns'].value_counts().values
```

```python
plt.pie(y,labels=x,colors=["red","green"],autopct="%0.2f%%")
plt.title("Sum of Run v/s Inning")
sns.set_style(style="dark")
plt.savefig("Sum of Run vs Inning.jpg")
plt.legend(["2nd Inning","1st Inning"])
plt.show()
```

# Q2

```python
plt.figure(figsize=(10,7))
sns.countplot(x="Opposition",data=dataset,color="red")
plt.xticks(rotation=20)
plt.ylabel("Count of Match")
plt.title("No of Opposition Team Face in ODI Match")
plt.yticks([i for i in range(0,30,5)])
sns.set_style(style="dark")
plt.savefig("No of Opposition Team Face in ODI Match.jpg")
plt.show()
```

# Q3

```python
plt.figure(figsize=(7,3))
sns.countplot(x="Dismissal",data=dataset[dataset['Opposition']=="Sri Lanka"],color = "red")
plt.ylabel("Count of Match")
plt.title("No of Dismissal in ODI Match Opp. Sri Lanka")
plt.yticks([i for i in range(0,16,2)])
sns.set_style(style="dark")
plt.savefig("No of Dismissal in ODI Match Opp. Sri Lanka.jpg")
plt.show()
```
